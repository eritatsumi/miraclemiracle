```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def testplot():
    x=np.arange(21)
    for i in range(2,9):
        y=i*0.3*np.sin(np.pi/10*x)
        plt.plot(x,y,'o-',label=str(i-1))
    
    plt.legend()
    plt.show()
```
